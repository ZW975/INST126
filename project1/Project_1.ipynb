```python
print("*****HelloÔºÅThis is your simple calclulator!*****")

#ask for user input and show the menu
print("Calclulator Menu")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Factorial")#A little addition I make
print("6. Power")
opt = int(input("Please input the type of calculation you want. Number only please: "))

#check input
#using a while loop to check if input is correct
#if not, while loop while not end
while opt < 1 or opt > 6:
    print("Wrong Input!")
    opt = int(input("Please input the type of calculation you want. Numbers only please: "))

#Addition
if opt == 1:
    print("Please enter the numbers you want to add.")
    
    add_in = float(input("Enter 0 to end inputs: "))
    if add_in == 0:
        print("Ends.")
#addition for more than 2 numbers, use 0 as the end number
#since nobody try to add 0 in additions...
#use while loop to add more than 2 numbers
    else:
        sum = 0
        while add_in != 0:
            sum += add_in
            add_in = float(input())

    result = sum

#Substraction
if opt == 2:
    print("Please enter the numbers you want to subtract.")
    
    #input
    num1 = float(input("Please input the first number: "))
    num2 = float(input("Please input the second number: "))

    #calculation
    result = num1 - num2

#Multiplication
if opt == 3:
    print("Please enter the numbers you want to multiply.")
    
    mul_in = float(input("Enter 0 to end inputs: "))
    if mul_in == 0:
        print("Ends.")
#Almost same as addition
#except it is multiplying
    else:
        out = 1
        while mul_in != 0:
            out *= mul_in
            mul_in = float(input())

    result = out

#Division
if opt == 4:
    print("Please enter the numbers you want to divide.")
    
    #input
    num1 = float(input("Please input the first number: "))
    num2 = float(input("Please input the second number: "))

    #calculation
    result = num1 / num2

#Factorial
#This is the one I wrote for EC4
if opt == 5:
    num = int(input("Please input a number: "))
    out = num
    
    while num != 1:
        out *= (num - 1)
        num -= 1
    
    result = out

#Power
#input
if opt == 6:
    num1 = float(input("Please input the number: "))
    num2 = float(input("Please input the power: "))

    #calculation
    result = num1**num2

#Result output
print("Result of calculation is", result)
```
